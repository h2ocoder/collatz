# Open Questions & Conjectures

This notebook documents confirmed patterns, new observations, conjectures arising from
the factorization and geometric investigations, and open questions for future research.

## Confirmed Patterns

The following patterns have been computationally verified in the preceding notebooks:

1. **Dropping Set structure**: Every integer n > 1 belongs to a unique Dropping Set_k, determined by the number of Collatz steps needed to first reach a value below n. Even numbers always have dropping time 1.

2. **Orbital Triple validity**: For every odd n > 1, the orbital triple (a, b, c) = (n^2 - d^2, 2dn, n^2 + d^2) satisfies a^2 + b^2 = c^2, forming a valid Pythagorean triple.

3. **Complex multiplier Re = 1/2**: The real part of the complex multiplier z_0 is always exactly 1/2 for all n > 1, placing all z_0 values on a vertical line in the complex plane.

4. **Stopping Point linearity**: Within each Stopping Class_k, the stopping points (d - n, d) lie on a finite set of parallel lines (the number of lines equals the stopping modulus).

5. **Proportional Power Ratio**: P(x) maps each orbit step to [0, 1), and the resulting polar plots exhibit structured spiral patterns dependent on the base.

## New Observations

Findings from the factorization investigation (Notebooks 03-04):

1. **Semiprime class prediction**: The Collatz class of a semiprime p*q is not fully determined by class(p) and class(q) alone. The same pair of factor classes can produce different composite classes, suggesting that more information (e.g., the specific primes, not just their classes) is needed.

2. **Near-trivial composites**: For n = 2^a * p, since n is even, its dropping set is always 1 (a single halving step). This is structurally trivial -- the factor of 2 dominates the first drop. The deeper question is whether class(p) influences the subsequent orbit structure after the initial halvings.

3. **Prime power divergence**: class(p^k) generally differs from class(p), and the relationship becomes more complex as k increases. There is no simple "power rule" for Collatz classes.

4. **Geometric separation**: In the complex plane, primes and composites may show statistical (but not deterministic) differences in their Im(z_0) distributions, reflecting different typical ratios of dropping destination to starting value.

5. **Factor signature clustering**: Numbers sharing the same factor class signature (tuples of their prime factors' dropping sets) show partial clustering in geometric space, but with significant overlap.

## Conjectures

### Conjecture 1 (Paper 1): Orbital Oddity Uniformity

**Statement**: All Dropping Orbits in Dropping Set_k share the same Orbital Oddity.

That is, if dropping_time(n) = dropping_time(m) = k, then the number of odd values in the dropping orbit of n equals the number of odd values in the dropping orbit of m.

**Status**: Computationally verified for all n up to tested limits. No counterexample found. This implies a rigid structure within each dropping set.

### Conjecture 2 (Paper 2): Stopping Point Modular Lines

**Statement**: All Stopping Points of Stopping Class_k fall on exactly Modulus_k parallel lines in the (x, y) = (d - n, d) plane.

The lines have a common slope determined by the class, and the number of distinct lines (the stopping modulus) follows a periodic pattern in the differences between consecutive class members.

**Status**: Computationally verified. The slope is given by stopping_line_slope(k) and the modulus by stopping_modulus_for_class(k).

### Conjecture 3 (New): Non-existence of Class Multiplication

**Statement**: There is no function f such that class(p * q) = f(class(p), class(q)) for all primes p, q.

The Collatz class of a product is not determined by the classes of its factors. Evidence: for multiple pairs (class_p, class_q), the composite class takes on more than one distinct value.

### Conjecture 4 (New): Weak Class Correlation

**Statement**: The mutual information between factor classes and composite class is positive but low relative to the entropy of the composite class.

Factor classes provide some (weak) statistical signal about the composite class, but fall far short of full prediction. The Collatz map interacts with multiplication in a fundamentally non-algebraic way.

## Open Questions

1. **Does prime class character predict composite class?**
   - The mutual information analysis suggests partial predictability. Can this be quantified more precisely? Is there an asymptotic bound on MI(class_factors; class_composite) as the range grows?

2. **Is there a class multiplication table?**
   - Evidence points to no, but perhaps a "probabilistic multiplication table" exists: for each (class_p, class_q), there may be a characteristic distribution over class(p*q) that converges as the primes grow.

3. **How do the geometric properties of primes differ from composites?**
   - The Im(z_0) distributions differ statistically. Is there a deeper reason? Does the distribution of dropping destinations d/n differ systematically between primes and composites?

4. **Do higher-order interactions matter?**
   - For numbers with three or more distinct prime factors, does the class depend on pairwise factor interactions, or only on the individual factor classes?

5. **What is the role of 2 in factorization?**
   - Powers of 2 trivially have dropping set 1. How does the 2-adic valuation of n (the largest power of 2 dividing n) interact with the Collatz class of the odd part of n?

6. **Can geometric clustering predict orbit length?**
   - If numbers with similar factor class signatures cluster in the complex plane, does their position predict total stopping time?

## Next Steps

1. **Scale up**: Extend the semiprime and prime power analyses to larger ranges (n up to 10,000 or 100,000) to test whether patterns are stable or artifacts of small samples.

2. **Probabilistic multiplication table**: For each (class_p, class_q) pair, compute the empirical distribution of class(p*q) over many primes and test for convergence.

3. **2-adic analysis**: Systematically study how the 2-adic valuation interacts with Collatz classification of the odd part.

4. **Machine learning**: Train a classifier to predict class(n) from prime factorization features. The accuracy ceiling reveals how much information factorization carries about Collatz behavior.

5. **Algebraic structure search**: Investigate whether Collatz classes form any group-like or ring-like structure under multiplication, even if not perfectly deterministic.

6. **Connect to number theory**: Relate the geometric clustering to known results about prime distribution (e.g., Dirichlet's theorem, Chebotarev density).
